 1. Import Libraries

 2. Load Dataset

    The file data.csv is loaded using read_csv().

 3. Data Inspection

    The following functions are used to understand the dataset:


*   head() → Displays first 5 rows
*   tail() → Displays last 5 rows

*   info() → Shows column details and data types

*   describe() → Provides statistical summary
*   isnull() → Checks for missing values


*   columns → Displays column names

    This helps in understanding structure and quality of data.

 4. Find Top 10 Products

    The dataset is grouped by Product Description.
    Total quantity sold is calculated using:

    groupby('Description')['Quantity'].sum()

    The top 10 products with highest sales quantity are selected.

 5. Data Visualization

    Two plots are created:

    **Line Plot**

    Shows top 10 products by total quantity sold.

    **Bar Plot**

    Clearly compares sales quantity of top 10 products.

 6. Conclusion

    The program analyzes sales data and identifies the top 10 most sold products, displaying them using graphical representations.

```python
#POOVIKA M - 24BAD087
#SCENARIO 1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

filepath="data.csv"
data=pd.read_csv(filepath,encoding="latin1")
print("POOVIKA M\n24BAD087\n")

print(data.head())
print(data.tail())
print(data.info())
print(data.describe())
print(data.isnull())

print(data.columns)

top_products = data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,8))
sns.lineplot(x=top_products.values,y=top_products.index)
plt.title("Top 10 Products by Sales Quantity")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product Name")
plt.show()
plt.figure(figsize=(10,8))
sns.barplot(x=top_products.values,y=top_products.index)
plt.title("Top 10 Products by Sales Quantity")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product Name")
plt.show()
print("success")


```

**CODE INTERPRETATION**

 1. Import Libraries

    pandas → Data handling

 2. Load Dataset

      info() → Data types and structure

      describe() → Statistical summary

      columns → Display column names

 4. Glucose Level Visualization

    Two plots are created for Glucose:

    Histogram with KDE → Shows distribution pattern.

    Boxplot → Identifies spread and possible outliers.

 5. Age Visualization

    Two plots are created for Age:

    Histogram with KDE → Shows age distribution.

    Boxplot → Shows variation and outliers.

 6. Correlation Analysis

    A heatmap is plotted using correlation matrix.
    This shows relationships between different health metrics such as glucose, age, BMI, etc.

```python
#SCENARIO 2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path="diabetes.csv";
data=pd.read_csv(file_path,encoding="latin1")
print("POOVIKA M\n24BAD087\n")

print(data.head())
print(data.tail())
print(data.info())
print(data.describe())
print(data.isnull())
print(data.columns)

# Visualization: Glucose

plt.figure(figsize=(10,6))
sns.histplot(data['Glucose'], bins=30, kde=True)
plt.title("Glucose Level Distribution (Histogram)")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=data['Glucose'])
plt.title("Glucose Level Distribution (Boxplot)")
plt.xlabel("Glucose Level")
plt.show()

# Visualization: Age

plt.figure(figsize=(10,6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title("Age Distribution (Histogram)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=data['Age'])
plt.title("Age Distribution (Boxplot)")
plt.xlabel("Age")
plt.show()

# Identify simple patterns

# Correlation between health metrics
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Health Metrics")
plt.show()
```

 4. Scatter Plots (Feature vs Price)

      Scatter plots are created to analyze relationships between house price and important features:


      Price vs Area → Shows how price changes with area.

      Price vs Bedrooms → Shows impact of number of bedrooms on price.

      Price vs Bathrooms → Shows impact of number of bathrooms on price.

 5. Correlation Analysis

```python
#SCENARIO 3

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path="Housing.csv"
my_data=pd.read_csv(file_path,encoding="latin1")
print("POOVIKA M\n24BAD087\n")

print(my_data.head())
print(my_data.tail())
print(my_data.info())
print(my_data.describe())
print(my_data.isnull())
print(my_data.columns)
plt.figure(figsize=(7,5))
plt.scatter(my_data['area'], my_data['price'], color='blue')
plt.title("Price vs Area")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(my_data['bedrooms'], my_data['price'], color='green')
plt.title("Price vs Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(my_data['bathrooms'], my_data['price'], color='red')
plt.title("Price vs Bathrooms")
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.show()


plt.figure(figsize=(10,8))
sns.heatmap(my_data[['price','area','bedrooms','bathrooms','stories','parking']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Housing Features")
plt.show()
```

**SCENARIO 4**

**CODE INTERPRETATION**

 1. Import Libraries

    pandas → Data handling

    matplotlib & seaborn → Data visualization

 2. Load Dataset

    The file marketing_campaign.csv is loaded using tab (\t) separator.
    The dataset is stored in my_data.
    Student details are printed for identification.

 3. Data Inspection

    The dataset is explored using:

    head() → First 5 rows

    tail() → Last 5 rows

    info() → Structure and data types

    describe() → Statistical summary

    isnull() → Check for missing values

    columns → Display column names

 4. Create New Feature (Age)

    A new column Age is created using:

    Age = 2026 − Year_Birth

    This converts birth year into current age.

 5. Bar Plots (Distribution Analysis)

    Bar plots are created for:

    Age Distribution → Shows number of customers in each age group.

    Income Distribution → Shows frequency of income values (top 10).

    Wine Spending (MntWines) → Shows spending pattern on wine (top 10).

    These plots help understand customer demographics and spending behavior.

 6. Box Plots (Spread & Outliers)

    Box plots are created for:

    Age → Shows distribution and outliers.

    Income → Shows variation in income levels.

    Wine Spending → Shows spread and extreme values.

    Box plots help identify variability and unusual values.

 7. Conclusion

    The program analyzes marketing campaign data, creates a new age feature, and visualizes customer age, income, and spending patterns using bar plots and box plots.

```python
#SCENARIO 4

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path="marketing_campaign.csv"
my_data=pd.read_csv(file_path,encoding="latin1",sep="\t")
print("POOVIKA M\n24BAD087\n")

print(my_data.head())
print(my_data.tail())
print(my_data.info())
print(my_data.describe())
print(my_data.isnull())
print(my_data.columns)
my_data['Age'] = 2026 - my_data['Year_Birth']

# BAR PLOTS

# Age bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['Age'].value_counts().index,
            y=my_data['Age'].value_counts().values)
plt.title("Age Distribution")
plt.show()

# Income bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['Income'].value_counts().index[:10],
            y=my_data['Income'].value_counts().values[:10])
plt.title("Income Distribution")
plt.show()

# Spending bar plot (Wine)
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['MntWines'].value_counts().index[:10],
            y=my_data['MntWines'].value_counts().values[:10])
plt.title("Spending Pattern (Wine)")
plt.show()

# BOX PLOTS

# Age box plot
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['Age'])
plt.title("Age Box Plot")
plt.show()

# Income bar plot
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['Income'])
plt.title("Income Box Plot")
plt.show()

# Spending bar plot (Wine)
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['MntWines'])
plt.title("Spending Box Plot")
plt.show()
```
